## Descripcion de problema No. 35
#### Este programa consiste en pedirle al usuario que ingrese varios números uno por uno. La idea es que el usuario pueda introducir todos los números que quiera, y cuando ya no desee continuar, simplemente escriba -1 para indicar que terminó (este número no se toma en cuenta).

#### Después de eso, el programa le pedirá que elija un dígito del 0 al 9. Con esa información, el programa revisará todos los números que el usuario ingresó anteriormente y contará cuántas veces aparece ese dígito dentro de ellos.

#### Al final, se mostrará el resultado de forma clara para que el usuario entienda fácilmente cuántas veces apareció el dígito que eligió.

#### En pocas palabras, el programa analiza una lista de números y busca cuántas veces se repite un dígito específico dentro de toda la secuencia.

```python
# 🎯 PROGRAMA: Contador de dígitos
# Autor: Alonso López García
# Descripción: Cuenta cuántas veces aparece un dígito en varios números

print("===================================")
print("🔢 CONTADOR DE DÍGITOS 🔢")
print("===================================")

# Inicializamos variables
numero = 0
contador_total = 0
digito = 0

# Variable para guardar todos los números en forma de texto (sin listas)
cadena_numeros = ""

print("\n👉 Ingresa números enteros uno por uno")
print("👉 Escribe -1 para terminar\n")

# 📥 Captura de números
while True:
    numero = int(input("Número: "))

    if numero == -1:
        break

    # Guardamos el número como texto separado por espacio
    cadena_numeros = cadena_numeros + str(numero) + " "

# 📥 Pedir el dígito
print("\n===================================")
digito = int(input("🔎 Ingresa un dígito (0-9): "))

# Convertimos el dígito a texto
digito_texto = str(digito)

# 🔍 Contar apariciones
i = 0

while i < len(cadena_numeros):

    if cadena_numeros[i] == digito_texto:
        contador_total = contador_total + 1

    i = i + 1

# 📊 Mostrar resultado
print("\n===================================")
print("📊 RESULTADO")
print("===================================")
print("El dígito", digito, "aparece", contador_total, "veces.")
print("===================================")
```
