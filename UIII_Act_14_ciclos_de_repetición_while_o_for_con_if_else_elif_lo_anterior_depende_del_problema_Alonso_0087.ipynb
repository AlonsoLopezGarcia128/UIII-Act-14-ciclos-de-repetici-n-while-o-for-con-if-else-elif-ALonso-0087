## Descripcion de problema No. 35
#### Este programa consiste en pedirle al usuario que ingrese varios números uno por uno. La idea es que el usuario pueda introducir todos los números que quiera, y cuando ya no desee continuar, simplemente escriba -1 para indicar que terminó (este número no se toma en cuenta).

#### Después de eso, el programa le pedirá que elija un dígito del 0 al 9. Con esa información, el programa revisará todos los números que el usuario ingresó anteriormente y contará cuántas veces aparece ese dígito dentro de ellos.

#### Al final, se mostrará el resultado de forma clara para que el usuario entienda fácilmente cuántas veces apareció el dígito que eligió.

#### En pocas palabras, el programa analiza una lista de números y busca cuántas veces se repite un dígito específico dentro de toda la secuencia.

## Prompt utilizado para crear codigo en Python (Mandado a la AI de NoteBook LM)
#### Actúa como un programador experto en Python y explica de forma clara (nivel principiante-intermedio).
#### Necesito que desarrolles un programa en Python que cumpla con lo siguiente:
*  El programa debe ser interactivo y permitir entrada de datos por parte del usuario.
*  El usuario debe ingresar números uno por uno.
*  La secuencia termina cuando el usuario ingrese el número -1 (este valor no se debe procesar).
*  Después, el usuario debe ingresar un dígito específico (del 0 al 9).
*  El programa debe contar cuántas veces aparece ese dígito dentro de todos los números ingresados.
*  Muestra el resultado de forma clara.

#### Además:

*  Haz el programa visual (usa menús, mensajes claros, emojis o decoraciones con texto).
*  Organiza el código en funciones.
*  No uses try/except ni validaciones complejas.
*  Añade comentarios en el código explicando cada parte.

```python
# --- ANALIZADOR DE DÍGITOS 'DIGIT-COUNT 2026' ---

def mostrar_encabezado():
    """Muestra un menú visual y decorativo al usuario."""
    print("=" * 40)
    print("✨ BIENVENIDO AL ANALIZADOR DE DÍGITOS ✨")
    print("=" * 40)
    print("Instrucciones:")
    print("1. Ingresa números uno por uno.")
    print("2. Para terminar la lista, ingresa -1.")
    print("-" * 40)

def capturar_secuencia():
    """Solicita números al usuario hasta que ingresa el centinela -1."""
    lista_numeros = [] # Lista vacía para almacenar las entradas [7, 8]

    while True: # Bucle infinito que se detiene con break [9, 10]
        entrada = input("🔢 Ingresa un número (o -1 para finalizar): ")

        # Verificamos si es el valor centinela para detenernos [11, 12]
        if entrada == "-1":
            break

        # Agregamos el número como string a la lista para facilitar el conteo [7]
        lista_numeros.append(entrada)

    return lista_numeros

def procesar_conteo(lista, digito_objetivo):
    """Cuenta las apariciones del dígito en todos los números de la lista."""
    total_apariciones = 0

    # Recorremos cada número almacenado en la lista [2, 13]
    for numero in lista:
        # Usamos el método count para contar el dígito dentro de cada cadena [5]
        total_apariciones += numero.count(digito_objetivo)

    return total_apariciones

def mostrar_resultado(digito, cantidad):
    """Presenta el informe final de forma clara y visual."""
    print("\n" + "📊 INFORME DE RESULTADOS")
    print("-" * 40)
    print(f"🔍 El dígito analizado fue: {digito}")
    print(f"✅ Aparece un total de: {cantidad} veces")
    print("-" * 40)
    print("🌟 ¡Gracias por usar LAB-PYTHON 2026! 🌟\n")

def ejecutar_programa():
    """Función principal que orquestra el flujo del programa."""
    mostrar_encabezado()

    # Paso 1: Obtener la lista de números
    datos_usuario = capturar_secuencia()

    # Validamos si la lista tiene datos antes de continuar [10, 14]
    if len(datos_usuario) == 0:
        print("⚠️ No ingresaste ningún número para procesar.")
    else:
        # Paso 2: Solicitar el dígito a buscar
        objetivo = input("\n🎯 ¿Qué dígito (0-9) deseas contar?: ")

        # Paso 3: Realizar el conteo
        resultado = procesar_conteo(datos_usuario, objetivo)

        # Paso 4: Mostrar el informe
        mostrar_resultado(objetivo, resultado)

# Iniciar la aplicación
ejecutar_programa()
```
